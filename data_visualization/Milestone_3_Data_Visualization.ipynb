## Milestone 3 – Data Visualization

## 1. Dataset Overview

```python
import os

parasitized_path = "cell_images/Parasitized"
uninfected_path = "cell_images/Uninfected"

print("Parasitized Images:", len(os.listdir(parasitized_path)))
print("Uninfected Images:", len(os.listdir(uninfected_path)))
```

### Observation

The dataset contains equal number of parasitized and uninfected cell images.
This balanced distribution helps prevent bias during model training.

## 2. Class Distribution Visualization

```python
import matplotlib.pyplot as plt

classes = ["Parasitized", "Uninfected"]
counts = [len(os.listdir(parasitized_path)),
          len(os.listdir(uninfected_path))]

plt.bar(classes, counts)
plt.title("Class Distribution")
plt.xlabel("Class")
plt.ylabel("Number of Images")
plt.show()
```

### Observation

The bar chart shows that both classes are equally represented in the dataset.
Balanced datasets improve classification performance.

## 3. Sample Image Visualization

```python
import cv2
import random
import matplotlib.pyplot as plt
import os

parasitized_path = "cell_images/Parasitized"
uninfected_path = "cell_images/Uninfected"

fig, axes = plt.subplots(1,4, figsize=(12,5))

# 2 Parasitized images
for i in range(2):
    img_name = random.choice(os.listdir(parasitized_path))
    img = cv2.imread(os.path.join(parasitized_path, img_name))
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    axes[i].imshow(img)
    axes[i].set_title("Parasitized")
    axes[i].axis("off")

# 2 Uninfected images
for i in range(2,4):
    img_name = random.choice(os.listdir(uninfected_path))
    img = cv2.imread(os.path.join(uninfected_path, img_name))
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    axes[i].imshow(img)
    axes[i].set_title("Uninfected")
    axes[i].axis("off")

plt.show()
```

### Observation

Parasitized cells show visible parasite structures inside red blood cells,
while uninfected cells appear clear and uniform.

## 4. Image Size Analysis

```python
import cv2
import os
import matplotlib.pyplot as plt

parasitized_path = "cell_images/Parasitized"
uninfected_path = "cell_images/Uninfected"

sizes = []

# Sirf 100 images check karenge fast run ke liye
for folder in [parasitized_path, uninfected_path]:
    for img_name in os.listdir(folder)[:100]:
        img = cv2.imread(os.path.join(folder, img_name))
        sizes.append(img.shape[0])  # height

plt.hist(sizes, bins=20)
plt.title("Image Height Distribution")
plt.xlabel("Image Height")
plt.ylabel("Count")
plt.show()
```

### Observation

The dataset contains images with varying dimensions.
Resizing will be required before training the deep learning model.

## 5. Pixel Intensity Distribution

```python
import cv2
import matplotlib.pyplot as plt
import os

parasitized_path = "cell_images/Parasitized"

# Ek sample image le rahe hain
img_name = os.listdir(parasitized_path)[0]
img = cv2.imread(os.path.join(parasitized_path, img_name))

# Gray convert
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.hist(gray.ravel(), bins=256)
plt.title("Pixel Intensity Distribution")
plt.xlabel("Pixel Value")
plt.ylabel("Frequency")
plt.show()
```

### Observation

Pixel intensity distribution helps understand brightness variation
across the dataset. Normalization may improve model performance.

```python
total_images = len(os.listdir(parasitized_path)) + len(os.listdir(uninfected_path))

parasitized_percent = (len(os.listdir(parasitized_path)) / total_images) * 100
uninfected_percent = (len(os.listdir(uninfected_path)) / total_images) * 100

print(f"Parasitized: {parasitized_percent:.2f}%")
print(f"Uninfected: {uninfected_percent:.2f}%")
```

```python
import numpy as np

heights = []
widths = []

for folder in [parasitized_path, uninfected_path]:
    for img_name in os.listdir(folder)[:100]:
        img = cv2.imread(os.path.join(folder, img_name))
        heights.append(img.shape[0])
        widths.append(img.shape[1])

print("Average Height:", np.mean(heights))
print("Average Width:", np.mean(widths))
```
